```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")

y = df_train["Purchase"]

df_train.drop(["Purchase"],axis=1,inplace=True)

df = pd.concat([df_train,df_test]).iloc[:, 2:]
```

```python
df.head(10)
```

```python
df.isna().sum()
```

```python
df = df.fillna(0)
```

```python
for i in df.columns:
    print(i+":",df[i].unique())
```

```python
from sklearn.preprocessing import LabelEncoder 

le = LabelEncoder()

df[df.columns] = df[df.columns].apply(lambda col: le.fit_transform(col))
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()

df = np.array(ohe.fit_transform(df).toarray(),dtype=np.str)
```

```python
from sklearn.preprocessing import StandardScaler

st = StandardScaler()
df = st.fit_transform(df)

df = pd.DataFrame(df)

df
```

```python
X = df.iloc[:550068]
X_val = df.iloc[550068:]

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3 , random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

model1 = RandomForestRegressor(random_state=0)

model1.fit(X_train,y_train)

y_pred1 = model1.predict(X_test)

from sklearn.metrics import mean_squared_error
from math import sqrt

print(sqrt(mean_squared_error(y_test,y_pred1)))
```

```python
from xgboost import XGBRegressor

model3 = XGBRegressor()

model3.fit(X_train,y_train)

y_pred3 = model3.predict(X_test)

print(sqrt(mean_squared_error(y_test,y_pred3)))
```

```python
from sklearn.linear_model import LinearRegression
model4 = LinearRegression()

model4.fit(X_train,y_train)

y_pred4 = model4.predict(X_test)

print(sqrt(mean_squared_error(y_test,y_pred4)))
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.regularizers import l2
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
model = Sequential()

model.add(Dense(100,input_dim = 94,kernel_initializer = "uniform",
                     kernel_regularizer = l2(0.1) , bias_regularizer = l2(0.1),
                     bias_initializer="zeros",activation="relu"))
model.add(Dense(50,kernel_initializer = "uniform",
                     kernel_regularizer = l2(0.1) , bias_regularizer = l2(0.1),
                     bias_initializer="zeros",activation="relu"))
model.add(Dense(1))

model.compile(optimizer="adam",loss = "mean_squared_error")

model.fit(X_train,y_train,batch_size=10 ,epochs=100)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)

print(np.sqrt(mean_squared_error(y_test,y_pred)))
```
