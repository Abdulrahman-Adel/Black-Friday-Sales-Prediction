```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")

y = df_train["Purchase"]

df_train.drop(["Purchase"],axis=1,inplace=True)
```

```python
df_train['User_Count']=df_train.groupby(['User_ID'])['User_ID'].transform('count')
df_train['Product_Count']=df_train.groupby(['Product_ID'])['Product_ID'].transform('count')
df_test['User_Count']=df_test.groupby(['User_ID'])['User_ID'].transform('count')
df_test['Product_Count']=df_test.groupby(['Product_ID'])['Product_ID'].transform('count')
```

```python
df = pd.concat([df_train,df_test])
df.head(10)
```

```python
l = dict(df["Product_Category_1"].value_counts())
m = []

for i in df["Product_Category_1"]:
    m.append(l[i])
    
df["Product_Category_Count"] = m  
```

```python
df.drop(["Product_Category_2","Product_Category_3"],axis=1,inplace=True)
```

```python
def df_transform_Age(x):
    if (x != '55+'):
        (min, max)=x.split('-')
        x = np.random.randint(int(min),int(max))
    else:
        x = np.random.randint(55,100)
    return x


df.Age=df.Age.apply(df_transform_Age)

for user in df.User_ID.unique():
    age=df.Age.loc[df.User_ID==user].values[0]
    df.Age.loc[df.User_ID==user]=age
```

```python
def Binned_Age(x):
    if int(x) <= 17 :
        x = "Teen Ages(0-17)"
    elif int(x) >= 18 and int(x)<25:
        x = 'Young Adult Ages(18-25)'
    elif int(x) >= 25 and int(x)<40:
        x = 'Middle Adult Ages(25-40)'
    elif int(x) >= 40 and int(x)<50:
        x = 'Adult Ages(40-50)'
    elif int(x) >= 50:
        x = 'Elderly(>50)'
    return x
df['Binned_Age'] = df.Age.apply(Binned_Age)
```

```python
df["Occupation"] = df["Occupation"].astype(str)
```

```python
df["Marital_Status"] = df["Marital_Status"].replace(1,"Maried").replace(0,"Single")
```

```python
def df_transform_Stay_In_Current_City_Years(x):
    if (x == '4+'):
        x = np.random.randint(4,10)
    return x
df.Stay_In_Current_City_Years=pd.to_numeric(df.Stay_In_Current_City_Years.apply(df_transform_Stay_In_Current_City_Years))
df
```

```python
from sklearn.preprocessing import LabelEncoder 

le = LabelEncoder()
onhot_cols = ['Occupation','City_Category']
categorical_columns = [x for x in df.columns if x not in onhot_cols]

categorical_columns
```

```python
df[categorical_columns] = df[categorical_columns].apply(lambda col: le.fit_transform(col))

df = pd.get_dummies(df, columns=onhot_cols)
```

```python
from sklearn.preprocessing import StandardScaler

st = StandardScaler()
df = st.fit_transform(df)
```
